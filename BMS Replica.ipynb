```python
class Movie:
    def __init__(self,rows,columns):
        self.rows=rows
        self.columns=columns
        self.Booking={}
        
        '''
          1 2 3 4 5
        1 s s s s s
        2 s s s s s
        '''
        
        
    def is_booking(self,row,col):
        seat_id=str(row)+str(col)
        if seat_id in self.Booking:
            return True
        else:
            return False
        
    def show_seats(self):
#         print("enter the row number:")
#         print("enter the column number: ")
        print("Here is your seat allocation: ")
        for row in range(0,self.rows+1):
            for column in range(0,self.columns+1):
                if row==0:
                    if column==0:
                        print(" ",end=" ")
                    else:
                        print(column,end=" ")
                else:
                    if column==0:
                        print(row,end=" ")
                    else:
                        if self.is_booking(row,column):
                            print("B",end=" ")
                        else:
                            print("s",end=" ")
            print()
            
    def book_ticket(self):
        print("Ticket counter")
        Booking_row=int(input("enter the row number you want to book :"))
        Booking_column=int(input("enter the column number you want to book :"))
        seat_id = str(Booking_row)+str(Booking_column)
        print("Here is your ticket price")
        total_seats=self.rows*self.columns
        if total_seats<=60:
            price=10
        elif total_seats>60:
            row_factor=self.rows//2
            if row_factor>=Booking_row:
                price=10
            else:
                price=8
        print("your ticket price is:",price)
        choice=int(input("Do you want to book?\n1. yes\n2. No\n"))
        if choice==1:
            print("we are booking your ticket")
            name=input("Enter your name: ")
            age=input("Enter your age: ")
            gender=input("Enter your gender: ")
            mob=input("Enter your mobile number: ")
            details = {'name':name,'gender':gender,'age':age,'mob':mob,'price':price}
            self.Booking[seat_id]=details
            print("your ticket is successfully booked")
        elif choice==2:
            print("cancelled booking")
            
    def show_details(self,row,col):
        seat_id=str(row)+str(col)
        if seat_id in self.Booking:
            detail=self.Booking[seat_id]
            print("Here is the details")
            print("name: ",detail["name"])
            print("age: ",detail["age"])
            print("gender: ",detail["gender"])
            print("mobile number: ",detail["mob"])
            print("The ticket is booked for price of: ",detail["price"])
        else:
            print(f"The ticket of selected row {row} and column {col} is not yet booked")
                     
    def statistics(self):
        purchased_tickets=len(self.Booking)
        print("Total purchased tickets : ",purchased_tickets)
        total_seats=self.rows*self.columns
        percentage_of_tickets=round((purchased_tickets/total_seats)*100,2)
        print("The percentage of booked tickets is: ",percentage_of_tickets)
        current_income=0
        for seat_id in self.Booking:
            details=m1.Booking[seat_id]
            price=details["price"]
            current_income+=price
        print("Current income is: ",current_income)
        if total_seats<=60:
            total_income=(self.rows*self.columns)*10
            print("Total income is: ",total_income)
        else:
            row_factor=self.rows//2
            front_rowseats=(row_factor*self.columns)
            last_rowseats=(self.rows-row_factor)*self.columns
            total_income=(front_rowseats)*10+(last_rowseats)*8
            print("total income is: ",total_income)
            
```
